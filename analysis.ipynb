```python
time = {}
for x in not_ready["LOG_ON_DATE_TIME"].unique():
    time[x] = {}
    date = x.split(" ")[0]
    for i in not_ready.loc[not_ready["LOG_ON_DATE_TIME"] == x]["REASON_CODE"]:
        time[x]["Date"] = date
        time[x]["Agent_ID"] = not_ready.loc[(not_ready["LOG_ON_DATE_TIME"] == x) & (not_ready["REASON_CODE"] == i)]["Agent_ID"].tolist()[0]
        time[x][i] = not_ready.loc[(not_ready["LOG_ON_DATE_TIME"] == x) & (not_ready["REASON_CODE"] == i)]["Ready"].tolist()[0]
```

```python
reasons = pd.DataFrame.from_dict(time, orient='index')
reasons.head()
```

```python
import datetime
new_date = []
for x in reasons["Date"]:
    new_date.append(datetime.datetime.strptime(x, '%m-%d-%Y').strftime('%m/%d/%y'))
```

```python
reasons.to_csv("Clean_Data/reasons_code.csv", index = False)
```
