## Clean Agent State Summary file

```python
# Load Agent State Summary 
Agent_State_Summary_df = pd.read_csv(agentstate_to_load)
Agent_State_Summary_df
```

```python
# Show column names for easier copy/paste for column removal
Agent_State_Summary_df.columns
```

```python
# Remove unnecessary columns
Agent_State_Summary_df.drop(columns = ['INTERNAL_CALS', 'EXTERNAL_OUTGOING_CALLS', 'NOT_READY', 'PER_NOT_READY',
        'WORK_TIME', 'PER_WORK_TIME', 'TALK_TIME', 'PER_TALK_TIME','AVG_TALK_TIME', 'TALK_TIME_OUT',
        'PER_TALK_TIME_OUT', 'EXT_OUT_CALLS_TALK_TIME', 'AVG_TALK_TIME_EXTERNAL','HOLD_TIME', 'PER_HOLD',
       'INTERNAL_CALLS_TIME', 'AVG_TALK_TIME_INTERNAL', 'PER_INTERNAL', 'BUSY_OTHER_TIME'], inplace=True)
Agent_State_Summary_df
```

```python
# Check group name values
Agent_State_Summary_df['GROUP_NAME'].value_counts()
```

```python
# Remove unnecessary group names
Agent_State_Summary_df = Agent_State_Summary_df.loc[(Agent_State_Summary_df['GROUP_NAME'] != 'SCH_CCM01.Cisco_Voice.defa.72419') & (Agent_State_Summary_df['GROUP_NAME'] != 'SCH_ONC_OVF_SG')]
Agent_State_Summary_df
```

```python
# Check group name values
Agent_State_Summary_df['GROUP_NAME'].value_counts()
```

```python
# Check that hours are between 8AM to 5PM
Agent_State_Summary_df['Hour'].value_counts()
```

```python
# Remove hours outside of 8AM to 5PM
Agent_State_Summary_df = Agent_State_Summary_df.loc[(Agent_State_Summary_df['Hour'] != '6:00 PM') & (Agent_State_Summary_df['Hour'] != '7:00 PM') & (Agent_State_Summary_df['Hour'] != '8:00 PM') & (Agent_State_Summary_df['Hour'] != '9:00 PM')& (Agent_State_Summary_df['Hour'] != '10:00 PM') & (Agent_State_Summary_df['Hour'] != '11:00 PM') & (Agent_State_Summary_df['Hour'] != '12:00 AM') & (Agent_State_Summary_df['Hour'] != '1:00 AM') & (Agent_State_Summary_df['Hour'] != '2:00 AM') & (Agent_State_Summary_df['Hour'] != '3:00 AM') & (Agent_State_Summary_df['Hour'] != '4:00 AM') & (Agent_State_Summary_df['Hour'] != '5:00 AM') & (Agent_State_Summary_df['Hour'] != '6:00 AM') & (Agent_State_Summary_df['Hour'] != '7:00 AM')]
Agent_State_Summary_df['Hour'].value_counts()
```
