```python
# Import dependencies
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Files to load
agentstate_to_load = "Working_Folder/Final_Project/call_center_data_2021/Agent_State_Summary_Hourly_2021.csv"
notready_to_load = "Working_Folder/Final_Project/call_center_data_2021/Agent_Not_Ready_Detail_2021.csv"
calltype_to_load = "Working_Folder/Final_Project/call_center_data_2021/Call_Type_Historical_Fields_Hourly_2021.csv"
```

## Clean Agent State Summary file

```python
# Load Agent State Summary 
Agent_State_Summary_df = pd.read_csv(agentstate_to_load)
Agent_State_Summary_df
```

### Remove unnecessary columns

```python
# Show column names for easier copy/paste for column removal
Agent_State_Summary_df.columns
```

```python
# Remove unnecessary columns
Agent_State_Summary_df.drop(columns = ['INTERNAL_CALS', 'EXTERNAL_OUTGOING_CALLS', 'NOT_READY', 'PER_NOT_READY',
        'WORK_TIME', 'PER_WORK_TIME', 'TALK_TIME', 'PER_TALK_TIME','AVG_TALK_TIME', 'TALK_TIME_OUT',
        'PER_TALK_TIME_OUT', 'HANDLED_CALLS_TALK_TIME', 'EXT_OUT_CALLS_TALK_TIME', 'AVG_TALK_TIME_EXTERNAL','HOLD_TIME', 'PER_HOLD',
       'INTERNAL_CALLS_TIME', 'AVG_TALK_TIME_INTERNAL', 'PER_INTERNAL', 'BUSY_OTHER_TIME'], inplace=True)
Agent_State_Summary_df
```

### Remove unnecessary values

```python
# Check group name values
Agent_State_Summary_df['GROUP_NAME'].value_counts()
```

```python
# Remove unnecessary rows
Agent_State_Summary_df = Agent_State_Summary_df.loc[(Agent_State_Summary_df['GROUP_NAME'] !=
                   'SCH_CCM01.Cisco_Voice.defa.72419') & (Agent_State_Summary_df['GROUP_NAME'] != 'SCH_ONC_OVF_SG')]
```

```python
# Check group name values
Agent_State_Summary_df['GROUP_NAME'].value_counts()
```

```python
# Check that hours are between 8AM to 5PM
Agent_State_Summary_df['Hour'].value_counts()
```

```python
# Remove hours outside of working hours (8AM to 5PM)
Agent_State_Summary_df = Agent_State_Summary_df.loc[(Agent_State_Summary_df['Hour'] != '6:00 PM')
            & (Agent_State_Summary_df['Hour'] != '7:00 PM') & (Agent_State_Summary_df['Hour'] != '8:00 PM')
            & (Agent_State_Summary_df['Hour'] != '9:00 PM')& (Agent_State_Summary_df['Hour'] != '10:00 PM')
            & (Agent_State_Summary_df['Hour'] != '11:00 PM') & (Agent_State_Summary_df['Hour'] != '12:00 AM')
            & (Agent_State_Summary_df['Hour'] != '1:00 AM') & (Agent_State_Summary_df['Hour'] != '2:00 AM')
            & (Agent_State_Summary_df['Hour'] != '3:00 AM') & (Agent_State_Summary_df['Hour'] != '4:00 AM')
            & (Agent_State_Summary_df['Hour'] != '5:00 AM') & (Agent_State_Summary_df['Hour'] != '6:00 AM')
            & (Agent_State_Summary_df['Hour'] != '7:00 AM')]
Agent_State_Summary_df['Hour'].value_counts()
```

### Group by pod

```python
# Create a function
def pod_name(x):
    if (x == 'CN_SCH_SPECC_EN_SG') or (x == 'CN_SCH_SPECC_SP_SG'):
        return 'POD 3'
    elif (x == 'CN_SCH_SPECC_POD_2_SG') or (x == 'CN_SCH_SPECC_SP_POD_2_SG'):
        return 'POD 2'
    else:
        return 'POD 1'

# Create new column to assign values
Agent_State_Summary_df['POD_NAME'] = Agent_State_Summary_df['GROUP_NAME'].apply(pod_name)
Agent_State_Summary_df
```

```python
# Check pods
Agent_State_Summary_df['POD_NAME'].value_counts()
```

### Change dtypes

```python
# Check dtypes
print(Agent_State_Summary_df.dtypes)

Agent_State_Summary_df
```

```python
# Change dtypes
Agent_State_Summary_df['Date'] = Agent_State_Summary_df['Date'].astype('string')
Agent_State_Summary_df['Date'] = pd.to_datetime(Agent_State_Summary_df['Date'])
Agent_State_Summary_df['GROUP_NAME'] = Agent_State_Summary_df['GROUP_NAME'].astype('string')
Agent_State_Summary_df['PER_AVAIL'] = Agent_State_Summary_df['PER_AVAIL'].str.rstrip('%').astype('float') / 100.0
Agent_State_Summary_df['POD_NAME'] = Agent_State_Summary_df['POD_NAME'].astype('string')

# Change LOGGED_ON_TIME hh:mm:ss format to seconds
def get_seconds(time_str):
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)

Agent_State_Summary_df['LOGGED_ON_TIME'] = Agent_State_Summary_df['LOGGED_ON_TIME'].astype('string')
Agent_State_Summary_df['LOGGED_ON_TIME'] = Agent_State_Summary_df['AVAILABLE_TIME'].apply(get_seconds)


# Change AVAILABLE_TIME hh:mm:ss format to seconds
Agent_State_Summary_df['AVAILABLE_TIME'] = Agent_State_Summary_df['AVAILABLE_TIME'].astype('string')
Agent_State_Summary_df['AVAILABLE_TIME'] = Agent_State_Summary_df['AVAILABLE_TIME'].apply(get_seconds)


# Change HANDLED_CALLS_TIME hh:mm:ss format to seconds
Agent_State_Summary_df['HANDLED_CALLS_TIME'] = Agent_State_Summary_df['HANDLED_CALLS_TIME'].astype('string')
Agent_State_Summary_df['HANDLED_CALLS_TIME'] = Agent_State_Summary_df['HANDLED_CALLS_TIME'].apply(get_seconds)


# Change AVG_HANDLE_TIME hh:mm:ss format to seconds
Agent_State_Summary_df['AVG_HANDLE_TIME'] = Agent_State_Summary_df['AVG_HANDLE_TIME'].astype('string')
Agent_State_Summary_df['AVG_HANDLE_TIME'] = Agent_State_Summary_df['AVG_HANDLE_TIME'].apply(get_seconds)


print(Agent_State_Summary_df.dtypes)
Agent_State_Summary_df
```

```python
# Remove unnecessary columns
Call_Type_df.drop(columns = ['SL','ASA','ATT','ASSIGNED_FROM_Q','AWT', 'Flow Out', 'Return', 'Calls Error', 'Other',
                             'PER_QUEUED', 'PER_ABAN','AV_ABAN_DELAY', 'SHORT_CALLS',], inplace=True)
Call_Type_df
```

### Change dtypes

```python
# Change dtypes
Call_Type_df['CALL_TYPE_NAME'] = Call_Type_df['CALL_TYPE_NAME'].astype('string')
Call_Type_df['DATE'] = pd.to_datetime(Call_Type_df['DATE'])


# Change AHT hh:mm:ss format to seconds
Call_Type_df['AHT'] = Call_Type_df['AHT'].astype('string')

def get_seconds(time_str):
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)

Call_Type_df['AHT'] = Call_Type_df['AHT'].apply(get_seconds)


# Change MAX_CALL_WAIT_TIME hh:mm:ss format to seconds
Call_Type_df['MAX_CALL_WAIT_TIME'] = Call_Type_df['MAX_CALL_WAIT_TIME'].astype('string')

def get_seconds(time_str):
    hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)

Call_Type_df['MAX_CALL_WAIT_TIME'] = Call_Type_df['MAX_CALL_WAIT_TIME'].apply(get_seconds)


print(Call_Type_df.dtypes)

Call_Type_df
```
